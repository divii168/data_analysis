```python
# Set Seaborn style
sns.set_style("darkgrid")

# Histogram (Distribution of numerical variables)
df.hist(figsize=(12, 8), bins=30, edgecolor='black')
plt.suptitle("Distribution of Numerical Features", fontsize=16)
plt.show()
```

```python
#2. Boxplot (Detect outliers)
# ===========================
plt.figure(figsize=(12, 6))
sns.boxplot(data=df)
plt.xticks(rotation=45)
plt.title("Boxplot for Outlier Detection")
plt.show()
```

```python
# PIE CHART - Air Quality Distribution
# ==============================
if 'Air Quality' in df.columns:
    air_quality_counts = df['Air Quality'].value_counts()  # Count unique values

    # Plot Pie Chart
    plt.figure(figsize=(8, 8))
    plt.pie(
        air_quality_counts, 
        labels=air_quality_counts.index, 
        autopct='%1.1f%%', 
        colors=sns.color_palette("pastel"), 
        startangle=140
    )
    
    plt.title("Air Quality Distribution")
    plt.show()
```
